```python
# High dimensionality dataset (784 features) on classification problem

import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
import miceforest as mf
from feature_selection import FeatureSelector
import feature_selection
import importlib

df = pd.read_csv('high_dimensionality.csv')
X = df.drop(columns=['label'])
y = df['label']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
fs = FeatureSelector(X_train, y_train, task_type='classification')

fs.methods['classification'].remove('jmim')
```

```python
fs.run_selection()

results_table = fs.get_results_table()
count_table = fs.get_count()

print("Results Table:")
print(results_table)
print("\nCount Table:")
print(count_table)

all_selected = fs.all_features_selected()
print("\nFeatures selected by all methods:")
print(all_selected)

selected_by_at_least_3 = fs.selected_by_at_least(3)
print("\nFeatures selected by at least 3 methods:")
print(selected_by_at_least_3)
```
