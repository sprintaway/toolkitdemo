```python
import numpy as np
import pandas as pd
from sklearn.datasets import make_blobs
from outlier_detection import OutlierDetector
import matplotlib.pyplot as plt

# Generate synthetic data
n_samples = 2000
n_outliers = int(0.1 * n_samples)
n_features = 4

X, _ = make_blobs(n_samples=n_samples - n_outliers, centers=3, n_features=n_features, random_state=42)

outliers = np.random.uniform(low=-10, high=10, size=(n_outliers, n_features))

X = np.vstack([X, outliers])

columns = [f'feature_{i}' for i in range(n_features)]
df = pd.DataFrame(X, columns=columns)

y = np.zeros(n_samples)
y[-n_outliers:] = 1

od = OutlierDetector(df, y)

od.run_selection()

selected_by_four = od.selected_by_at_least(4)

true_outliers = set(range(n_samples - n_outliers, n_samples))
detected_outliers = set(selected_by_four)
false_outliers = detected_outliers - true_outliers

fig, ax = od.pcaplot(selected_by_four)

ax.legend()

plt.show()

print(f"Number of true outliers: {n_outliers}")
print(f"Number of points detected as outliers by at least 4 methods: {len(selected_by_four)}")
print(f"Number of false outliers: {len(false_outliers)}")

results_table = od.get_results_table()
print("\nResults Table:")
print(results_table)

count_table = od.get_count()
print("\nCount Table:") 
print(count_table)
```

```python
import numpy as np
import pandas as pd
from sklearn.datasets import make_moons
from outlier_detection import OutlierDetector
import matplotlib.pyplot as plt

n_samples = 2000
n_outliers = int(0.1 * n_samples)  

X, _ = make_moons(n_samples=n_samples - n_outliers, noise=0.1, random_state=42)

outliers = np.random.uniform(low=-1, high=2, size=(n_outliers, 2))

X = np.vstack([X, outliers])

columns = ['feature_1', 'feature_2']
df = pd.DataFrame(X, columns=columns)

y = np.zeros(n_samples)
y[-n_outliers:] = 1

od = OutlierDetector(df, y)

od.run_selection()

all_selected = od.selected_by_at_least(3)

true_outliers = set(range(n_samples - n_outliers, n_samples))
detected_outliers = set(all_selected)
false_outliers = detected_outliers - true_outliers

fig, ax = od.pcaplot(all_selected)

ax.legend()

plt.show()

print(f"Number of true outliers: {n_outliers}")
print(f"Number of points detected as outliers by at least 3 methods: {len(all_selected)}")
print(f"Number of false outliers: {len(false_outliers)}")

results_table = od.get_results_table()
print("\nResults Table:")
print(results_table)

count_table = od.get_count()
print("\nCount Table:") 
print(count_table)
```

```python
import numpy as np
import pandas as pd
from sklearn.datasets import make_gaussian_quantiles
from outlier_detection import OutlierDetector
import matplotlib.pyplot as plt

n_samples = 2000
n_outliers = int(0.1 * n_samples)  

X, _ = make_gaussian_quantiles(n_samples=n_samples - n_outliers, n_features=2, n_classes=1, random_state=42)

outliers = np.random.uniform(low=-4, high=4, size=(n_outliers, 2))

X = np.vstack([X, outliers])

columns = ['feature_1', 'feature_2']
df = pd.DataFrame(X, columns=columns)

y = np.zeros(n_samples)
y[-n_outliers:] = 1

od = OutlierDetector(df, y)

od.run_selection()

all_selected = od.selected_by_at_least(3)

true_outliers = set(range(n_samples - n_outliers, n_samples))
detected_outliers = set(all_selected)
false_outliers = detected_outliers - true_outliers

fig, ax = od.pcaplot(all_selected)

ax.legend()

plt.show()

print(f"Number of true outliers: {n_outliers}")
print(f"Number of points detected as outliers by at least 3 methods: {len(all_selected)}")
print(f"Number of false outliers: {len(false_outliers)}")

results_table = od.get_results_table()
print("\nResults Table:")
print(results_table)

count_table = od.get_count()
print("\nCount Table:")
print(count_table)
```
