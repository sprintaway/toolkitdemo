```python
import pandas as pd
from sklearn.model_selection import train_test_split
import distribution_comparison
import proposed_train_test_split
import hdbscan

df = pd.read_csv('resale_flat_prices.csv')
X = df.drop(columns=['price_per_sqft'])
y = df['price_per_sqft']

X_train_normal, X_test_normal, y_train_normal, y_test_normal = train_test_split(X, y, test_size=0.2, random_state=0)

print("Comparing distributions after normal train-test split:")
distribution_comparison.compare_train_test_distributions(X_train_normal, X_test_normal)

```

```python
# Proposed train-test split
X_train_proposed, X_test_proposed, y_train_proposed, y_test_proposed = proposed_train_test_split.proposed_train_test_split(
    X, y, test_size=0.2, min_cluster_size=5, random_state=0, task='regression'
)

print("\nComparing distributions after proposed train-test split:")
distribution_comparison.compare_train_test_distributions(X_train_proposed, X_test_proposed)
```
